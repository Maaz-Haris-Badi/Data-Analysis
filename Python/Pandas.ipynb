Merge, Joins, Concatenate

Inner Join = Things which are in both the dataframes, just like intersection
Outer join = All the things in both data frames, just like Union.
Left Join = Includes all the things in the left data frame as well as similar things from the rigth data frame
Right Join = Includes all the things in the right data frame as well as similar things from the left data frame

```python
import pandas as pd

df1 = pd.read_csv(r"D:\Projects\Data Analysis\Python\Data\LOTR.csv")
df2 = pd.read_csv(r"D:\Projects\Data Analysis\Python\Data\LOTR 2.csv")

#normal merge (Inner Join)
df1.merge(df2, on = ["FellowshipID", "FirstName"])
#outer Join
df1.merge(df2, how = "outer")
```
