```python
import pandas as pd 
df = pd.read_csv(r"F:\Projects\Data Analysis\Customer Analysis\Data\customer_info.csv")
# removing duplicates
df.drop_duplicates(inplace=True)

# cleaning and standardiizing the loyality tier column
df["loyalty_tier"] = df["loyalty_tier"].str.replace("gold", "GOLD")
df["loyalty_tier"] = df["loyalty_tier"].str.replace(" GOLD ", "GOLD")
df["loyalty_tier"] = df["loyalty_tier"].str.replace("gld", "GOLD")
df["loyalty_tier"] = df["loyalty_tier"].str.replace("Silver", "SILVER")
df["loyalty_tier"] = df["loyalty_tier"].str.replace("sllver", "SILVER") 
df["loyalty_tier"] = df["loyalty_tier"].str.replace("bronze", "BRONZE") 
df["loyalty_tier"] = df["loyalty_tier"].str.replace("brnze", "BRONZE")
df["loyalty_tier"].unique() 
```

```python
#checking unique values in region column
df["region"].unique()
#region column is clean and standardiized so no changes needed
```

```python
#cleaning and standardizing the gender column
df["gender"] = df["gender"].str.replace("Female", "FEMALE")
df["gender"] = df["gender"].str.replace("femle", "FEMALE")
df["gender"] = df["gender"].str.replace("Male", "MALE")
df["gender"] = df["gender"].str.replace("male", "MALE")
df["gender"].unique()
```

```python
df.to_csv(r"F:\Projects\Data Analysis\Customer Analysis\Data\customer_info_cleaned.csv", index=False)
```

```python
import pandas as pd
sd = pd.read_csv(r"F:\Projects\Data Analysis\Customer Analysis\Data\sales_data.csv")

#droping duplicates
sd.drop_duplicates(inplace=True)
```

```python
# checking unique values in the payment method column
sd['payment_method'].unique()

# Standardizing payment method column
sd['payment_method'] = sd['payment_method'].str.replace("credit card", "Credit Card")
sd['payment_method'] = sd['payment_method'].str.replace("bank transfr", "Bank Transfer")

sd['payment_method'].unique()
```

```python
# checking unique values in delivery status column
sd['delivery_status'].unique()

#cleaning and standardizing delivery status column
sd['delivery_status'] = sd['delivery_status'].str.replace("delivered", "DELIVERED")
sd['delivery_status'] = sd['delivery_status'].str.replace("Delivered", "DELIVERED")
sd['delivery_status'] = sd['delivery_status'].str.replace("delrd", "DELIVERED")
sd['delivery_status'] = sd['delivery_status'].str.replace("Cancelled", "CANCELLED")
sd['delivery_status'] = sd['delivery_status'].str.replace(" DELAYED", "DELAYED")
sd['delivery_status'] = sd['delivery_status'].str.replace("Delayed", "DELAYED")
sd['delivery_status'] = sd['delivery_status'].str.replace("delyd", "DELAYED")

sd['delivery_status'].unique()
```

```python
# converting discount_applied column to float and filling NaN with 0.0 
sd['discount_applied'] = sd['discount_applied'].astype('float')
sd['discount_applied'] = sd['discount_applied'].fillna(0.0)
```

```python
#cleaning and standardizing the quantity column
sd['quantity'] = sd['quantity'].str.replace('three', '3')
sd['quantity'] = sd['quantity'].str.replace('five', '5')

sd['quantity'] = sd['quantity'].astype('float')
sd['quantity'].unique()
```

```python
sd['quantity'] = sd['quantity'].fillna(sd['quantity'].median())
sd['quantity'].unique()

sd
```

```python
sd.to_csv(r"F:\Projects\Data Analysis\Customer Analysis\Data\sales_data_cleaned.csv", index=False)
```
